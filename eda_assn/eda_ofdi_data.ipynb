# Univariate Analysis

##Numerical feature analysis

```python
plt.figure(figsize=(10, 6))
sns.distplot(df_copy['equity'], kde=True)
plt.xlabel('Equity in million USD')
plt.title("Distribution of Equity")
plt.tight_layout()
plt.show()
```

This right skewness suggests that the highest number of equity are bought at a lower rate but there are some few exceptions which go beyond 100 million USD.

```python
plt.figure(figsize=(10, 6))
sns.distplot(df_copy['loan'], kde=True)
plt.xlabel('Loan in million USD')
plt.title("Distribution of Loan")
plt.tight_layout()
plt.show()
```

Again we can see the extreme right skewness which implies that the most amount of loan that is issued is in small amount, with some exception of large sum of money given that drives the higher range

```python
plt.figure(figsize=(10, 6))
sns.distplot(df_copy['guarantee_issued'], kde=True)
plt.xlabel('Guarantee Issued in million USD')
plt.title("Distribution of Guarantee Issued")
plt.tight_layout()
plt.show()
```

The gurantee issued variable, while also being skewed shows the most amount of skewness implying that the outliers in this feature have a bigger impact to the overall feature.

```python
plt.figure(figsize=(10, 6))

sns.histplot(df_copy['total'], kde=True, bins=20)
plt.title("Distrubution of Total Financial Commitment in million USD")
plt.xlabel("Total financial commitment")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

The right skewness in the total field shows that it is common characterisitic of financial data to show some exceptionally large values, which might influence the feature as a whole.

```python
columns = ['loan', 'equity', 'guarantee_issued', 'total']

fig, ax = plt.subplots(2, 2, figsize=(20, 20))

for i, col in enumerate(columns):
  row, col_index = divmod(i, 2)
  sns.boxplot(df_copy[col], ax=ax[row, col_index])
  ax[row, col_index].set_title(f"Box Plot of {col.capitalize()}")
  ax[row, col_index].set_ylabel(f"{col.capitalize()}")

plt.tight_layout()
plt.show()
```

## Categorical Feature analysis

```python
df_copy['major_activity'].unique()
```

```python
plt.figure(figsize=(12, 8))

df_copy['major_activity'].value_counts().plot(kind='barh', width=0.5)
plt.gca().invert_yaxis()
plt.title('Frequency of Investments in Major Activites')
plt.xlabel('Major Activities')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

As the data suggests, the most amount of investments happen in the **Financial, Insurance and Business Services**, after that **Wholesale, Retail Trade, Restraunts and Hotels** takes up the most sum of money. Including manufacturing with the above 2, it takes up the more than double the sum of the money in all the other sectors.

```python
plt.figure(figsize=(6, 6))

df_copy['overseas_country'].value_counts()[:5].plot(kind='pie', autopct='%1.1f%%')
plt.title('Countries with most amount of Investment from India')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

The pie chart suggests that the most amount of investments is taken up by **UAE** and **USA**. The combined takes up even more than 50% of the top 5 countries with most Indian Investors.

```python
plt.figure(figsize=(8, 8))

df_copy['whether_jv_wos'].value_counts().plot(kind='bar')
plt.title('Distribution of Investment in JV/WOS')
plt.xlabel('Types of Overseas Direct Investment')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

As the bar plot suggests, there is significant difference between the number of investments happening in Wholly Owned Subsidaries(WOS) and Joint Ventures(JV), with WOS taking over JVs by almost double the amount.

# Bivariate Analysis

```python
indian_parties = df_copy.groupby('name_of_indian_party')['total'].sum().sort_values(ascending=False)
indian_parties.head(10)
```

```python
indian_parties.shape
```

```python
plt.figure(figsize=(10, 6))

indian_parties[:5].plot(kind='barh', width=0.8)
plt.gca().invert_yaxis()
plt.title('Financial Commitment by Indian Parties')
plt.xlabel('Total Financial Commitment')
plt.ylabel('Name of Indian Parties')

plt.tight_layout()
plt.show()
```

As the graph suggests, the indian parties which invest the most in overseas country is the **Biocon Biologics Limited** with over 800 million USD. The next in line is **Vedanta Limited** with over 700 million USD. Rest of the parties lies even below the half mark made by these 2 parties alone.

```python
most_invested_country = df_copy.groupby('overseas_country')['total'].sum().sort_values(ascending=False)
most_invested_country.head(10)
```

```python
most_invested_country.shape
```

```python
plt.figure(figsize=(11, 6))

most_invested_country[:5].plot(kind='barh', width=0.5)
plt.gca().invert_yaxis()
plt.title('Country with Investment amount')
plt.xlabel('Investment amount')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

As the bar graph suggests, the investment with more money has been done in **UK** followed by **Mauritius**. This shows that the number of investments differ from the amount of investment that was done in these countries.

```python
total_investments = df_copy['total'].sum()

percent_loan = ((df_copy['loan'].sum())/total_investments) * 100
percent_equity = ((df_copy['equity'].sum())/total_investments) * 100

print(f"Percent Loan: {percent_loan:.2f}%, Percent Equity: {percent_equity:.2f}%")
```

```python
categories = ['Loan', 'Equity', 'Gurantee Issued']
values = [percent_loan, percent_equity, 100-percent_equity-percent_loan]

plt.figure(figsize=(6, 6))

plt.pie(values, labels=categories, autopct='%1.1f%%')
plt.title('Percentage Distribution of Financial Commitments')

plt.tight_layout()
plt.show()
```

The above pie chart suggests that the amount of Investments have been done on basis of Gurantee Issued followed by the Loan given.
